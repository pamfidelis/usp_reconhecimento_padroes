# Dataset covid-19

### From: https://www.kaggle.com/sudalairajkumar/novel-corona-virus-2019-dataset?select=COVID19_open_line_list.csv

```python
import pandas as pd

df = pd.read_csv("dataset/dataset_base_congelado.csv", low_memory=False)

# para mostrar toda a base
pd.set_option('display.max_columns', None)

df
```

## Tratamento de features

```python
# remover colunas nulas
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

#remover missing values
df = df.dropna(axis=0, subset=['outcome'])

# remover linhas duplicacoes
df = df.drop_duplicates()

df
```

```python
# remoção de campos não usados
df.drop(columns =['admin3',
                      'admin2',
                      'admin1',
                      'country_new',
                      'admin_id',
                      'data_moderator_initials',
                      'notes_for_discussion',
                      'location',
                      'additional_information',
                      'source',
                      'sequence_available',
                      'geo_resolution',
                      'wuhan(0)_not_wuhan(1)',
                      'reported_market_exposure'], inplace = True)

df
```

## Tratamento de dados

```python
# transformando dados em binário

df.sex = df.sex.map({ 'male' : 1, 'famale' : 0})

df.outcome = df.outcome.map({ 'discharged': 0, 'discharge': 0, 'died': 1, 'death': 1 })


# validando Nan

df.latitude.fillna(0, inplace=True)
df.longitude.fillna(0, inplace=True)


df
```

```python
# tratamento de idade
d#f['age'] = df['age'].astype(int)
df_age = pd.DataFrame(df['age'])

def transforms_age_young(valor):
    if valor <= 19:
        return 1
    else:
        return 0

def transforms_age_adult(valor):
    if valor > 19 and valor <= 59:
        return 1
    else:
        return 0

def transforms_age_elderly(valor):
    if valor > 59 :
        return 1
    else:
        return 0

df_age['age_binario_young'] = df_age['age'].map(transforms_age_young)  
df_age['age_binario_adult'] = df_age['age'].map(transforms_age_adult)
df_age['age_binario_elderly'] = df_age['age'].map(transforms_age_elderly)


df_age.drop(columns =['age'], inplace=True)

df2 = df.loc[:,:'ID']
df3 = df.loc[:,'sex':]

df =df2.append([df_age, df3]) 


df
```

```python
# indexar por outcome
df = df.set_index('outcome')
df = df.loc[[1,0]]

df
```
